checking the stock avaliability

```python
for sp in soup.find_all('li',class_ ="col-xs-6 col-sm-4 col-md-3 col-lg-3"):

  print("--"*20)
  avaliable_stock = sp.find('p',class_ = 'instock availability').text.strip()
  print(avaliable_stock)
```

```python
for sp in soup.find_all('li',class_ ="col-xs-6 col-sm-4 col-md-3 col-lg-3"):

  print("--"*20)
  book_link = sp.find_all('a')[-1].get('href')
  title = sp.find_all('a')[-1].get('title')
  img_link = sp.find('img').get('src')
  rating = sp.find('p').get('class')[-1]
  price = sp.find('p',class_ = 'price_color').text[2:]
  avaliable_stock = sp.find('p',class_ = 'instock availability').text.strip()

  print(title," | ",book_link," | ",img_link," | ",rating," | ",price," | ",avaliable_stock)
```

```python
data = []
for sp in soup.find_all('li',class_ ="col-xs-6 col-sm-4 col-md-3 col-lg-3"):

  book_link = sp.find_all('a')[-1].get('href')
  title = sp.find_all('a')[-1].get('title')
  img_link = sp.find('img').get('src')
  rating = sp.find('p').get('class')[-1]
  price = sp.find('p',class_ = 'price_color').text[2:]
  avaliable_stock = sp.find('p',class_ = 'instock availability').text.strip()

  data.append([title,book_link,img_link,rating,price,avaliable_stock])
```
