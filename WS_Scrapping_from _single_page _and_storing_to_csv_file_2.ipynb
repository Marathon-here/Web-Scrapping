```python
#all the quotes one after the other

#storing all the data into a list
data = []
for sp in soup.find_all('div',class_='quote'):
  #print(sp)
  qoute = sp.find('span',class_ = 'text').text[1:-1]
  author = sp.find('small',class_='author').text
  author_id = sp.find('a').get('href')

  tags = []
  for tag in sp.find_all('a',class_='tag') :
    tags.append(tag.text)

  tags = ','.join(tags) #to be seperated by commas in a line
  data.append([quote,author,author_id,tags])
  print(qoute)
  print(author)
  print(author_id)
  print(tags)
  print()
  print("--"*20)
  print()
```
