```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm  #for better presentation of pages progress
import pandas as pd
```

```python
req=requests.get('https://quotes.toscrape.com/')
```

```python
soup = BeautifulSoup(req.text,'html.parser')
```

```python
#all the quotes one after the other

#storing all the data into a list
data = []
for sp in soup.find_all('div',class_='quote'):
  #print(sp)
  quote = sp.find('span',class_ = 'text').text[1:-1]
  author = sp.find('small',class_='author').text
  author_id = sp.find('a').get('href')

  tags = []
  for tag in sp.find_all('a',class_='tag') :
    tags.append(tag.text)

  tags = ','.join(tags) #to be seperated by commas in a line
  data.append([quote,author,author_id,tags])
  print(quote)
  print(author)
  print(author_id)
  print(tags)
  print()
  print("--"*20)
  print()
```

#### once the Soup is prepared for multiple pages we can use the similar approch of single page for extracting contents of everypage

```python
data = []
for page in tqdm(range(1,11)):
  link = 'https://quotes.toscrape.com/page/'+str(page)
  response = requests.get(link)
  soup = BeautifulSoup(response.text, 'html.parser')  # soup is prepared of multiple pages

  #build the same logic to single webpage scrapper for soup

  for sp in soup.find_all('div',class_='quote'):
  #print(sp)
    quote = sp.find('span',class_ = 'text').text[1:-1]
    author = sp.find('small',class_='author').text
    author_id = sp.find('a').get('href')

    tags = []
    for tag in sp.find_all('a',class_='tag') :
      tags.append(tag.text)

    tags = ','.join(tags) #to be seperated by commas in a line
    data.append([quote,author,author_id,tags])
    print(quote)
    print(author)
    print(author_id)
    print(tags)
    print()
    print("--"*20)
    print()
```

```python
len(data)
```

formating data into dataset or dataframe **formate**

```python
df = pd.DataFrame(data, columns=['quote','author','author_id','tags'])
```
